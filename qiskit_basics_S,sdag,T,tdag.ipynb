# Main

```python
#Pauli-X

circuit=QuantumCircuit(1)
# circuit.x(0)
circuit.ry(pi/4, 0)
print(circuit)
# circuit.draw(output= 'mpl')
```

```python
#Pauli-Y

circuit=QuantumCircuit(2)
circuit.ry(pi/4,[0,1])
circuit.y(1)
# circuit.ry(pi/4, 0)
print(circuit)
# circuit.draw(output= 'mpl')
```

```python
#Pauli-Z

circuit=QuantumCircuit(2)
circuit.ry(pi/4,[0,1])
circuit.z(1)
# circuit.ry(pi/4, 0)
print(circuit)
# circuit.draw(output= 'mpl')
```

```python
#Degrees to Radian Converter
deg= float(input("Enter degree:"))
val=deg*(pi/180)
print(val)
```

```python
#Hadamard Gates
circuit = QuantumCircuit(3)
# circuit.x(0)
# circuit.h([0,1,2])
# circuit.s(1)
# circuit.sdg(1)
# circuit.s(1)
# circuit.s(1)
# circuit.h(1)

circuit.x(0)
circuit.h(1)
# circuit.h(2)
# circuit.x(2)


circuit.ry(val,2)
# circuit.rz(val,[0,2])
# circuit.r(val,)
# circuit.h(2)
# circuit.rz(pi/3,2)
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend= simulator ).result()
statevector=result.get_statevector()
# plot_histogram(result.get_counts())
array_to_latex(statevector, prefix = '\\text{statevector = } \n')
```

```python
plot_bloch_multivector(statevector)
```

```python
circuit.measure_all()
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator, shots=1000000).result()
plot_histogram(result.get_counts())
```
